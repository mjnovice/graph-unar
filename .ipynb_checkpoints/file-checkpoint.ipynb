```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
import pytesseract
from pytesseract import Output

def get_data(im):
    return pytesseract.image_to_data(im, output_type=Output.DICT)

def get_comp(data,i):
    comp = {
                     'text':data['text'][i],
                     'x_start':data['left'][i],
                     'x_end': data['left'][i]+data['width'][i],
                     'y_start':data['top'][i],
                     'y_end':data['top'][i]+data['height'][i],
            }
    return comp

def get_texts(img):
    data = get_data(img)
    texts=[]
    n = len(data['text'])
    print(data['text'])
    for i in range(n):
        if len(data['text'][i].strip())==0:
            continue
        comp = get_comp(data,i)
        texts.append(comp)
    return texts

def get_bargraph_boundaries(img):
    print(img.shape[:2])
    x_start,y_start = 0,0
    x_end, y_end = img.shape[:2]
    texts = get_texts(img)
    
    #find the topmost text and the start of the y-axis
    topmost_ylimit = y_end + 1
    x_axis_ttls_ylimit = 0
    y_axis_ttls_xlimit = x_end + 1

    #find the topmost text first
    for text in texts:
        if topmost_ylimit > text['y_end']:
            topmost_ylimit = text['y_end']
    
    #find the y_axis_ttls_xlimit
    for text in texts:
        if topmost_ylimit < text['y_end'] and text['x_end'] > y_axis_ttls_xlimit:
            y_axis_ttls_xlimit = text['x_end']
    
    #find the x_axis_ttls_ylimit
    for text in texts:
        if topmost_ylimit < text['y_end'] and text['text'].isnumeric() and text['y_end'] > x_axis_ttls_ylimit:
            x_axis_ttls_ylimit = text['y_end']

    
    x_ttls = []
    y_ttls = []
    err = 2
    for text in texts:
        if text['y_end']>=x_axis_ttls_ylimit:
            x_ttls.append(text)
    
    for text in texts:
        if text['x_end']<=y_axis_ttls_xlimit:
            y_ttls.append(text)

    #boundary of the xaxis and yaxis labels
    yaxis=None
    xaxis=None
    print(texts)
    for text in texts:
        if xaxis is None and text['text'].isnumeric():
            xaxis = text
        if yaxis is None and not text['text'].isnumeric():
            if text['y_start']>topmost['y_end']:
                yaxis = text
    y_start = max(y_start,topmost['y_end'])
    x_start = max(x_start,yaxis['x_end'])
    y_end = min(y_end, xaxis['y_start'])
    return {'x_start':x_start,'y_start':y_start,'y_end':y_end,'x_end':x_end}


def get_x_y_ttls(img):
    print(img.shape[:2])
    x_start,y_start = 0,0
    x_end, y_end = img.shape[:2]
    texts = get_texts(img)
    
    #find the topmost text and the start of the y-axis
    topmost_ylimit = y_end + 1
    x_axis_ttls_ylimit = 0
    y_axis_ttls_xlimit = 0

    #find the topmost text first
    for text in texts:
        if topmost_ylimit > text['y_end']:
            topmost_ylimit = text['y_end']
    
    #find the y_axis_ttls_xlimit
    for text in texts:
        if topmost_ylimit < text['y_end'] and not text['text'].isnumeric() and text['x_end'] > y_axis_ttls_xlimit:
            y_axis_ttls_xlimit = text['x_end']
    
    #find the x_axis_ttls_ylimit
    for text in texts:
        if topmost_ylimit < text['y_end'] and text['text'].isnumeric() and text['y_end'] > x_axis_ttls_ylimit:
            x_axis_ttls_ylimit = text['y_end']

    
    x_ttls = []
    y_ttls = []
    err=5
    for text in texts:
        if abs(text['y_end']-x_axis_ttls_ylimit)<=err:
            x_ttls.append(text)
    
    for text in texts:
        if abs(text['x_end']-y_axis_ttls_xlimit)<=err:
            y_ttls.append(text)
    return x_ttls,y_ttls


def get_bar_texts(img,bb):
    im=img.copy()
    bargraph=im[bb['y_start']:bb['y_end'],bb['x_start']:bb['x_end']]
    im = cv2.bitwise_not(bargraph)
    data = pytesseract.image_to_data(im, output_type=Output.DICT,config='--psm 6')
    texts = []
    n = len(data['text'])
    for i in range(n):
        if not data['text'][i].isnumeric():
            print(data['text'][i])
            continue
        comp = get_comp(data,i)
        #offset
        comp['x_start'] += bb['x_start']
        comp['y_start'] += bb['y_start']
        comp['x_end'] += bb['x_start']
        comp['y_end'] += bb['y_start']
        texts.append(comp)
    return texts

def not_white(color):
    white_threshold=250
    return color[0]<white_threshold and color[1]<white_threshold and color[2]<white_threshold

def get_bar_coords(bar_texts,im):
    img = im.copy()
    bar_coords=[]
    for text in bar_texts:
        tx,ty = text['x_start'],text['y_start']
        bx,by = text['x_end'],text['y_end']
        bound = {
                     'x_start':tx,
                     'x_end': bx,
                     'y_start':ty,
                     'y_end':by,
        }
        #cv2.rectangle(img,(tx,ty),(bx,by),(0,255,0),2)
        print("origin",img[ty,tx],img[by,bx])
        while not_white(img[ty,tx]):
            tx-=1
        tx+=1
        bound['x_start']=tx
        tx,ty = text['x_start'],text['y_start']
        while not_white(img[ty,tx]):
            ty-=1
        ty+=1
        bound['y_start']=ty
        bx,by = text['x_end'],text['y_end']
        while not_white(img[by,bx]):
            bx+=1
        bx-=1
        bound['x_end']=bx
        bx,by = text['x_end'],text['y_end']
        while not_white(img[by,bx]):
            by+=1
        by-=1
        bound['y_end']=by
        cv2.rectangle(im,(bound['x_start'],bound['y_start']),(bound['x_end'],bound['y_end']),(0,255,0),2)

        print(img[ty,tx],img[by,bx])
        bar_coords.append(bound)
    return bar_coords
#get_bar_coords(img[bb['y_start']:bb['y_end'],bb['x_start']+13:bb['x_end']])

img=cv2.imread('barchart.png')
#bb = get_bargraph_boundaries(img)
x_axis_titles, y_axis_titles = get_x_y_ttls(img)
print(x_axis_titles, y_axis_titles)
#bar_texts = get_bar_texts(img, bb)
#print(bt)
#print(get_bar_coords(bt,img))
plt.rcParams['figure.figsize'] = [20, 15]
plt.imshow(img)
```
